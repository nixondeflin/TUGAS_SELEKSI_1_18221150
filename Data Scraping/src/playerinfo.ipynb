```python
import pandas as pd
from selenium import webdriver
from selenium.webdriver.common.by import By

from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
```

```python
driver = webdriver.Chrome()
driver.get('https://liquipedia.net/valorant/VCT/2023/Pacific_League/Statistics')
```

```python
playerNames = driver.find_elements(By.XPATH,'//tbody/tr/td[1]/a')

list = []

for each in playerNames:
    list.append(each.text)

print(list)
```

```python
dfpName= pd.DataFrame(list,columns=["Player Nickname"])
dfpName
```

```python
new =[]
for i in range(len(list)):
    temp = []
    string = 'https://liquipedia.net/valorant/' + list[i]
    driver.get(string)
    pName = driver.find_element(By.XPATH,'//div/div[2]/div[1]/div[4]/div[2]')
    temp.append(pName.text)
    try:
        birth = driver.find_element(By.XPATH,"//*[contains(text(), '(age')]")
        temp.append(birth.text)
    except:
        pass
    new.append(temp)
```

```python
dfpinfo = pd.DataFrame(new,columns=["Player Name","Birth"])
dfpinfo
```

```python
dfnew = pd.concat([dfpName,dfpinfo],axis=1,join='inner')
dfnew
```

```python
dffinal = pd.concat([dfnew,])
```
